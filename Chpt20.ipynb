### 20.1 Run Snippets 20.1 and 20.2 with timeit. Repeat 10 batches of 100 executions. What is the minimum elapsed time for each snippet?

```python
import timeit

test_code_20_1 = '''
dict0 = {'a': ['1', '2'], 'b': ['+', '*'], 'c': ['!', '@']}
for a in dict0['a']:
	for b in dict0['b']:
		for c in dict0['c']:
			print({'a': a, 'b': b, 'c': c})
'''

print(min(timeit.Timer(test_code_20_1).repeat(repeat=10, number=100)))
```

```python
test_code_20_2 = '''
from itertools import product
dict0 = {'a': ['1', '2'], 'b': ['+', '*'], 'c': ['!', '@']}
jobs = (dict(zip(dict0, i)) for i in product(*dict0.values()))
for i in jobs: print(i)
'''

print(min(timeit.Timer(test_code_20_2, setup='').repeat(repeat=10, number=100)))
```
